```python
import simpy
import numpy as np
```

```python
BETA_CLIENTES_1 = 240/60 # 1/LAMBDA
BETA_CLIENTES_2 = 120/60 # 1/LAMBDA
BETA_CLIENTES_3 = 360/60 # 1/LAMBDA

def obtener_tipo():
    prob = np.random.uniform(0,1)
    if prob < .1 :
        return 1
    elif prob < .8:
        return 2
    return 3

def arribo_de_clientes(env, cajero):
    cliente_id = 0
    while True:
        tipo = obtener_tipo()
        c = cliente(env, cliente_id, cajero, tipo)
        env.process(c)
        tiempo = env.now
        if tiempo < 120:
            beta = BETA_CLIENTES_1
        elif tiempo < 120+180:
            beta = BETA_CLIENTES_2
        else:
            beta = BETA_CLIENTES_3
            
        cliente_siguiente = np.random.exponential(beta)                
        # Espero al cliente
        yield env.timeout(cliente_siguiente)
        #print('LegÃ³ un cliente en tiempo: ', env.now)
        cliente_id += 1

def obtener_tiempo_de_uso(tipo):
    if tipo == 1:
        return np.random.uniform(1,7)
    elif tipo == 2:
        return np.random.uniform(1,3)
    return np.random.uniform(1,5)
tiempos_de_espera = []
personas_en_la_fila = [0]
def cliente(env, cliente_id, cajero, tipo):
    """Customer arrives, is served and leaves."""
    arrive = env.now
    print('%7.4f %s: Here I am' % (arrive, cliente_id))
    personas_en_la_fila.append(personas_en_la_fila[-1]+1)
    with cajero.request() as cj:
        yield cj
        personas_en_la_fila.append(personas_en_la_fila[-1]-1)
        # Wait for the counter or abort at the end of our tether
        mi_turno = env.now
        espera = mi_turno - arrive
        tiempos_de_espera.append(espera)
        print('%7.4f %s: estoy usando el cajero, soy del tipo %s ' % (mi_turno, cliente_id, tipo))
        t = obtener_tiempo_de_uso(tipo)
        yield env.timeout(t)
        print('%7.4f %s: Finished' % (env.now, cliente_id))
        
        


# Setup and start the simulation
env = simpy.Environment()

# Start processes and run
cajero = simpy.Resource(env, capacity=1)
env.process(arribo_de_clientes(env, cajero))
unidades_de_tiempo = 60 # 540 
env.run(until = unidades_de_tiempo)
print('tiempo de espera maximo: ',max(tiempos_de_espera))
print(personas_en_la_fila)
print('largo maximo de la cola: ',max(personas_en_la_fila))
```

```python
"""
Movie renege example

Covers:

- Resources: Resource
- Condition events
- Shared events

Scenario:
  A movie theatre has one ticket counter selling tickets for three
  movies (next show only). When a movie is sold out, all people waiting
  to buy tickets for that movie renege (leave queue).

"""
import collections
import random

import simpy

BETA = 45

def obtener_tipo_solicitud():
    prob = np.random.uniform(0,1)
    if prob < .7 :
        return 1
    elif prob < .9:
        return 2
    return 3

def obtener_tiempo_de_procesamiento(tipo):
    if tipo == 1:
        return np.random.uniform(60, 180)
    elif tipo == 2:
        return np.random.uniform(120, 360)
    return np.random.uniform(200, 800)

def obtener_servidor_round_robin(granja, solicitud_id):
    id_servidor =  solicitud_id % len(granja)
    print("servidor: ", id_servidor)
    return granja[id_servidor]

def procesar_solicitud(env, solicitud_id, servidor, tipo):
    """Customer arrives, is served and leaves."""
    arrive = env.now
    print('%7.4f %s: Nueva olicitud' % (arrive, solicitud_id))
    
    with servidor.request() as srv:
        yield srv
        mi_turno = env.now
        print('%7.4f %s: Procesando solicitud del tipo %s ' % (mi_turno, solicitud_id, tipo))
        t = obtener_tiempo_de_procesamiento(tipo)
        yield env.timeout(t)
        print('%7.4f %s: Solicitud procesada' % (env.now, solicitud_id))

def arribo_de_solicitudes(env, func ,granja):
    """Create new *moviegoers* until the sim time reaches 120."""
    id_solicitud = 0
    while True:
        
        servidor = obtener_servidor_round_robin(granja, id_solicitud)
        solicitud = procesar_solicitud(env, id_solicitud, servidor, obtener_tipo_solicitud())
        env.process(solicitud)
        yield env.timeout(np.random.exponential(BETA))
        
        id_solicitud += 1    
  
```

```python
# Setup and start the simulation
env = simpy.Environment()

servidor_1 = simpy.Resource(env, capacity=1)
servidor_2 = simpy.Resource(env, capacity=1)
servidor_3 = simpy.Resource(env, capacity=1)
servidor_4 = simpy.Resource(env, capacity=1)
servidor_5 = simpy.Resource(env, capacity=1)

GRANJA = [servidor_1, servidor_2, servidor_3, servidor_4, servidor_5]

# Start processes and run
env.process(arribo_de_solicitudes(env, obtener_servidor_round_robin, GRANJA))
unidades_de_tiempo = 10000 # 540 
env.run(until = unidades_de_tiempo)
```
